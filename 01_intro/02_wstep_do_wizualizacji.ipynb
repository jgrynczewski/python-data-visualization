# Wstęp do wizualizacji danych

## 1. Wprowadzenie - Czym jest wizualizacja danych?

**Wizualizacja danych** to graficzne przedstawienie informacji i danych. Używając elementów wizualnych takich jak wykresy, grafy i mapy, wizualizacja danych dostarcza nam przystępny sposób na zobaczenie i zrozumienie trendów, wartości odstających i wzorców w danych.

### Dlaczego wizualizacja jest ważna?

1. **Szybsze zrozumienie** - Nasz mózg przetwarza obrazy 60 000 razy szybciej niż tekst
2. **Odkrywanie wzorców** - Łatwiej zauważyć trendy i anomalie na wykresie niż w tabeli
3. **Lepsze zapamiętywanie** - Wizualne przedstawienie informacji jest łatwiejsze do zapamiętania
4. **Efektywna komunikacja** - Jeden wykres może zastąpić długi opis słowny
5. **Podejmowanie decyzji** - Dane wizualne pomagają w szybszym podejmowaniu świadomych decyzji

### Przykłady zastosowań:

- **Biznes**: Dashboardy sprzedażowe, analiza trendów rynkowych
- **Nauka**: Prezentacja wyników badań, wizualizacja eksperymentów
- **Media**: Infografiki, wizualizacje danych statystycznych
- **Finanse**: Wykresy giełdowe, analiza portfela inwestycyjnego

### Przykład: Różnica między danymi a wizualizacją

Wyobraź sobie tabelę z 1000 wierszami sprzedaży. Trudno zauważyć trend.

Ale wykres liniowy pokazujący sprzedaż w czasie? Od razu widać:
- Czy sprzedaż rośnie czy spada?
- Czy są jakieś sezonowe wzorce?
- Gdzie są wartości odstające?

**To jest siła wizualizacji!**

## 2. Dane wejściowe - Rodzaje danych

Zanim zaczniemy tworzyć wykresy, musimy zrozumieć, jakie rodzaje danych możemy wizualizować.

### Typy danych:

#### 1. Dane liczbowe (numeryczne)

**Ciągłe (continuous):**
- Mogą przyjmować dowolną wartość w zakresie
- Przykłady: temperatura (23.5°C), wzrost (175.3 cm), cena (99.99 zł)
- Wykresy: liniowe, punktowe, histogramy

**Dyskretne (discrete):**
- Mogą przyjmować tylko określone wartości (zazwyczaj liczby całkowite)
- Przykłady: liczba osób (5), liczba sprzedanych produktów (127)
- Wykresy: słupkowe, punktowe

#### 2. Dane kategoryczne

**Nominalne:**
- Nie mają naturalnej kolejności
- Przykłady: kolor (czerwony, niebieski), miasto (Warszawa, Kraków), dział (IT, Marketing)
- Wykresy: słupkowe, kołowe

**Porządkowe (ordynalne):**
- Mają naturalną kolejność
- Przykłady: poziom zadowolenia (niski, średni, wysoki), oceny (1, 2, 3, 4, 5)
- Wykresy: słupkowe

#### 3. Dane czasowe (temporal)
- Daty i czas
- Przykłady: 2024-01-15, 14:30:00
- Wykresy: liniowe, obszarowe

### Wczytywanie danych z pliku CSV

W Pythonie najczęściej używamy biblioteki **Pandas** do pracy z danymi. Pandas to potężne narzędzie do analizy i manipulacji danych.

````python
!pip install pandas
````

````python
# Import biblioteki pandas
import pandas as pd

# Wczytanie danych z pliku CSV
df_sprzedaz = pd.read_csv('data/sprzedaz.csv')

print("Dane zostały wczytane pomyślnie!")
print(f"Liczba wierszy: {len(df_sprzedaz)}")
````

### Podstawowa eksploracja danych

Przed tworzeniem wykresów zawsze warto przyjrzeć się danym. Oto najważniejsze funkcje:

````python
# head() - pokaż pierwsze 5 wierszy
print("Pierwsze 5 wierszy:")
print(df_sprzedaz.head())
````

````python
# info() - podstawowe informacje o danych
print("Informacje o zbiorze danych:")
print(df_sprzedaz.info())
````

````python
# describe() - statystyki opisowe dla kolumn numerycznych
print("Statystyki opisowe:")
print(df_sprzedaz.describe())
````

### Co możemy wywnioskować z tych informacji?

- `head()` pokazuje strukturę danych - jakie mamy kolumny i jak wyglądają przykładowe wartości
- `info()` mówi nam o typach danych i czy są braki danych (null values)
- `describe()` daje statystyki: średnią, minimum, maksimum, kwartyle

**Te informacje pomogą nam wybrać odpowiedni typ wykresu!**

## 4. Dobre praktyki wizualizacji danych

Tworzenie wykresów to nie tylko technika - to także sztuka komunikacji. Oto najważniejsze zasady:

### 4.1 Zasada KISS (Keep It Simple, Stupid)

**Prostota jest kluczem do skutecznej wizualizacji.**

✅ **Dobrze:**
- Jeden wykres = jedna główna idea
- Minimalna liczba elementów
- Czytelne czcionki
- Proste kolory

❌ **Źle:**
- Za dużo informacji na jednym wykresie
- Niepotrzebne efekty 3D
- Zbyt wiele kolorów
- Dekoracje odciągające uwagę

**Pamiętaj:** Jeśli wykres wymaga 5 minut wyjaśnień, to znaczy, że jest źle zaprojektowany!

### 4.2 Dobór odpowiedniego typu wykresu

Różne dane wymagają różnych wykresów. Oto przewodnik:

#### Porównanie wartości:
- **Wykres słupkowy** - porównanie kategorii (np. sprzedaż produktów)
- **Wykres słupkowy grupowany** - porównanie kategorii w grupach

#### Zmiany w czasie:
- **Wykres liniowy** - trendy czasowe (np. temperatura w ciągu roku)
- **Wykres obszarowy** - zmiana proporcji w czasie

#### Relacje między zmiennymi:
- **Wykres punktowy (scatter)** - korelacje (np. wiek vs pensja)
- **Heatmapa** - macierz korelacji

#### Rozkłady:
- **Histogram** - rozkład jednej zmiennej
- **Boxplot** - rozkład i wartości odstające
- **Violinplot** - szczegółowy rozkład w grupach

#### Proporcje:
- **Wykres kołowy** - TYLKO gdy masz 2-5 kategorii i sumują się do 100%
- **Wykres słupkowy ułożony** - lepsze rozwiązanie niż wykres kołowy

**Ważne:** Wykres kołowy jest często nadużywany. W większości przypadków wykres słupkowy jest lepszy!

### 4.3 Czytelne opisy osi i tytułów

Każdy wykres MUSI mieć:

1. **Tytuł** - co pokazuje wykres?
   - ✅ "Sprzedaż produktów w 2023 roku"
   - ❌ "Wykres"

2. **Opisy osi** - co przedstawia każda oś?
   - ✅ "Temperatura [°C]", "Miesiąc"
   - ❌ "X", "Y"

3. **Jednostki** - zawsze podawaj jednostki!
   - ✅ "Cena [zł]", "Czas [s]"
   - ❌ "Cena", "Czas"

4. **Legenda** - gdy masz wiele serii danych
   - Umieść ją w miejscu, które nie zakrywa danych
   - Użyj jasnych nazw

**Zasada:** Wykres powinien być zrozumiały bez czytania towarzyszącego tekstu!

### 4.4 Palety kolorów - dostępność i czytelność

Kolory są ważne, ale muszą być używane mądrze:

#### Zasady doboru kolorów:

1. **Nie używaj więcej niż 5-7 kolorów** na jednym wykresie
2. **Używaj kolorów ze znaczeniem:**
   - Czerwony = negatywne, niebezpieczne, spadki
   - Zielony = pozytywne, bezpieczne, wzrosty
   - Niebieski = neutralny, standard

3. **Pomyśl o daltonizmie:**
   - ~8% mężczyzn ma problem z rozróżnianiem czerwieni i zieleni
   - Używaj palet colorblind-friendly (np. viridis, cividis)

4. **Kontrast:**
   - Upewnij się, że kolory są wyraźnie rozróżnialne
   - Sprawdź wykres w skali szarości

5. **Spójność:**
   - Ten sam kolor = ta sama kategoria we wszystkich wykresach
   - Nie zmieniaj kolorów między slajdami

#### Polecane palety:
- **Sekwencyjna** (od jasnego do ciemnego): dla danych uporządkowanych (np. temperatury)
- **Dywergentna** (dwa kolory z białym centrum): dla danych z punktem środkowym (np. korelacje)
- **Jakościowa** (różne kolory): dla kategorii bez kolejności

### 4.5 Unikanie "chartjunk"

**Chartjunk** to wszystkie elementy wykresu, które nie niosą informacji, a tylko odwracają uwagę.

#### Przykłady chartjunk (usuń to!):

❌ **Niepotrzebne efekty 3D**
- Wykresy kołowe 3D są trudniejsze do odczytania
- Słupki 3D zniekształcają proporcje

❌ **Zbędne dekoracje**
- Tła z gradientami
- Obrazki w tle
- Ozdobne ramki

❌ **Za dużo linii siatki**
- Siatka powinna pomagać, nie przeszkadzać
- Często wystarczy tylko pozioma siatka

❌ **Zbyt ozdobne czcionki**
- Używaj prostych, czytelnych czcionek (Arial, Helvetica, sans-serif)

#### Zasada: **Data-Ink Ratio**

Maksymalizuj stosunek "tuszu" użytego na dane do całkowitego "tuszu" wykresu.

**Każdy element wykresu powinien coś wnosić. Jeśli go usuniesz i wykres będzie równie czytelny - usuń go!**

### Podsumowanie dobrych praktyk

1. ✅ **Prostota** - jeden wykres, jeden przekaz
2. ✅ **Odpowiedni typ** - dobierz wykres do rodzaju danych
3. ✅ **Czytelne opisy** - tytuł, osie, jednostki, legenda
4. ✅ **Mądre kolory** - przemyślana paleta, dostępność
5. ✅ **Bez zbędności** - każdy element ma cel

**Złota zasada:** Twój wykres powinien być zrozumiały dla kogoś, kto widzi go po raz pierwszy, bez wyjaśnień!

### 2.3. JupyterLab - następna generacja Jupyter

**JupyterLab** to nowocześniejsze, rozszerzone środowisko pracy - następna generacja interfejsu Jupyter. Jest to **rekomendowane narzędzie** do pracy z notebookami.

#### 2.3.1. Czym jest JupyterLab?

JupyterLab to zaawansowane środowisko IDE dla:
- Notebooków Jupyter (`.ipynb`)
- Edytora tekstu
- Terminala
- Podglądu plików (CSV, JSON, obrazy, PDF)
- Konsoli interaktywnej

**Główne różnice: Jupyter Notebook vs JupyterLab**

| Cecha | Jupyter Notebook | JupyterLab |
|-------|------------------|------------|
| **Interfejs** | Prosty, jeden notebook na zakładkę | Wielopanelowy, IDE-like |
| **Praca z wieloma plikami** | Osobne zakładki przeglądarki | Zakładki i panele w jednym oknie |
| **Edytor tekstu** | Nie | Tak, wbudowany |
| **Terminal** | Nie | Tak, wbudowany |
| **Drag & drop** | Ograniczony | Pełna obsługa (komórki, pliki) |
| **Podgląd plików** | Ograniczony | CSV, JSON, Markdown, obrazy, PDF |
| **Rozszerzalność** | Ograniczona | Bogaty system rozszerzeń |
| **Wydajność** | Dobra | Lepsza dla dużych projektów |
| **Status** | Klasyczny, legacy | Aktywnie rozwijany, przyszłość |

## 8. Magic Commands - polecenia magiczne

**Magic commands** to specjalne komendy IPython dostępne w Jupyter Notebook. Zaczynają się od `%` (line magic) lub `%%` (cell magic).

### 8.1. Podstawowe magic commands

| `%load` | Line | Wczytuje kod z pliku |

### 8.2. %matplotlib inline - wyświetlanie wykresów

**UWAGA:** W nowszych wersjach JupyterLab i Jupyter Notebook (5.0+), backend `inline` jest często **ustawiany domyślnie**, więc wykresy wyświetlają się automatycznie bez tej komendy.

Niemniej jednak, dobrą praktyką jest **jawne ustawienie** `%matplotlib inline` na początku notebooka, ponieważ:
- Zwiększa czytelność kodu (jasno pokazuje intencję)
- Zapewnia kompatybilność ze starszymi wersjami
- Pozwala łatwo zmienić backend (np. na `notebook`, `widget`)

**Dostępne backendy:**
- `%matplotlib inline` - Statyczne wykresy inline (domyślne w JupyterLab)
- `%matplotlib widget` - Interaktywne wykresy (wymaga ipympl)
- `%matplotlib notebook` - Interaktywne wykresy (starsza wersja, deprecated)

**Użyj go raz na początku notebooka dla pewności!**

`````python
# Magic command: %matplotlib inline
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

# Prosty przykład
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Wykres funkcji sinus')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()

print("Wykres pojawił się bezpośrednio nad tym tekstem!")
`````

### 8.3. %%timeit - pomiar wydajności

Mierzy czas wykonania kodu (wykonuje wiele razy i oblicza średnią).

`````python
# Liniowa wersja - mierzy tylko tę linię
%timeit [str(i**2) for i in range(100)]
`````

`````python
%%timeit
# Komórkowa wersja - mierzy całą komórkę
# Musi być pierwszą linią komórki!

x = np.linspace(0, 10, 1000)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y)
plt.close(fig)  # Zamykamy aby nie wyświetlać
`````

`````python
%matplotlib inline

import random
from matplotlib import pyplot as plt
import numpy as np

# Wykres sinusa z szumem
x1 = [x*0.01 for x in range(0, 628)]
y1 = [np.sin(x*0.01) + random.gauss(0, 0.1) for x in range(0, 628)]
plt.plot(x1, y1)

# Wykres cosinusa+ random.gauss(0, 0.1)
x2 = [x*0.5 for x in range(0, round(63/5))]
y2 = [np.cos(x*0.5) for x in range(0, round(63/5))]
plt.plot(x2, y2, 'o-')

plt.show()
`````

## 17. Tworzenie prezentacji (Slides)

Jupyter Notebook pozwala na tworzenie profesjonalnych prezentacji!

### 17.1. Jak utworzyć prezentację?

**Krok 1:** Włącz toolbar do slajdów
- Menu: **View → Cell Toolbar → Slideshow**

**Krok 2:** Dla każdej komórki wybierz typ slajdu:
- **Slide** - Nowy slajd (przesuwany w prawo)
- **Sub-Slide** - Pod-slajd (przesuwany w dół)
- **Fragment** - Element pojawia się po kliknięciu
- **Skip** - Pomijany w prezentacji
- **Notes** - Notatki prelegenta (niewidoczne w prezentacji)

**Krok 3:** Eksportuj do slajdów

```bash
# Podstawowy eksport
$ jupyter nbconvert notebook.ipynb --to slides

# Eksport i natychmiastowe uruchomienie w przeglądarce
$ jupyter nbconvert notebook.ipynb --to slides --post serve
```

### 17.2. Zaawansowane opcje slajdów

```bash
# Z niestandardowym motywem i opcjami
$ jupyter nbconvert --to slides index.ipynb --reveal-prefix=reveal.js \
    --SlidesExporter.reveal_theme=serif \
    --SlidesExporter.reveal_scroll=True \
    --SlidesExporter.reveal_transition=none
```

### 17.3. GitHub Pages z Jupyter Slides

```bash
# Dodaj reveal.js jako submodule
$ git submodule add https://github.com/hakimel/reveal.js.git reveal.js

# Wygeneruj slajdy
$ jupyter nbconvert --to slides index.ipynb --reveal-prefix=reveal.js
```

**Dostępne motywy:** beige, black, blood, league, moon, night, serif, simple, sky, solarized, white

**Typy przejść:** none, fade, slide, convex, concave, zoom

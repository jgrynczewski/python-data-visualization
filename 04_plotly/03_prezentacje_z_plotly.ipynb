# Prezentacje z Wykresami Plotly - Kompletny Przewodnik

**Pytanie:** Mam wykresy w Plotly/Dash - jak zrobić z nich prezentację?

**Odpowiedź:** Jest kilka sposobów - od prostego exportu do PowerPointa po interaktywne prezentacje HTML!

## Metoda 1: Export do PNG/SVG → PowerPoint/Google Slides

### Jak to działa:
1. Zapisz wykres jako **statyczny obraz** (PNG, SVG, PDF)
2. Wstaw obraz do PowerPoint/Google Slides/Keynote
3. **Standardowa prezentacja** - jak zawsze

### ✅ Plusy:
- **Najprostsze** - każdy zna PowerPoint
- Działa wszędzie (offline, email, PDF)
- Pełna kontrola nad layoutem slajdów
- Można dodać animacje, tekst, inne grafiki

### ❌ Minusy:
- Wykresy są **statyczne** (brak hover, zoom, pan)
- Utrata interaktywności Plotly
- Trzeba exportować każdy wykres osobno

### Kiedy używać:
- **Typowe prezentacje** biznesowe
- Wysyłasz PDF przez email
- Potrzebujesz slajdów z tekstem + wykresy
- Publiczność nie musi interagować z wykresami

---

### Kod - Export do PNG

**WAŻNE:** Wymaga zainstalowania `kaleido` (backend do eksportu):
```bash
pip install kaleido
```

````python
import plotly.express as px
import pandas as pd

# Przykładowy wykres
df = px.data.gapminder().query("year == 2007")
fig = px.scatter(
    df, 
    x="gdpPercap", 
    y="lifeExp",
    size="pop", 
    color="continent",
    title="Life Expectancy vs GDP (2007)",
    width=1200,   # Szerokość dla prezentacji
    height=700    # Wysokość
)

# EXPORT do PNG
fig.write_image("wykres_prezentacja.png", width=1920, height=1080)  # Full HD

# EXPORT do SVG (wektor - lepsze skalowanie)
fig.write_image("wykres_prezentacja.svg")

# EXPORT do PDF
fig.write_image("wykres_prezentacja.pdf")

print("✅ Wykresy wyexportowane!")
print("Teraz: Otwórz PowerPoint → Insert → Picture → wykres_prezentacja.png")
````

### Porady do exportu PNG:

**Rozdzielczość:**
```python
# Dla slajdu Full HD (16:9)
fig.write_image("slide.png", width=1920, height=1080, scale=2)

# Dla druku (wysoka jakość)
fig.write_image("print.png", width=3000, height=2000, scale=3)

# Dla ekranu 4:3 (stare projektory)
fig.write_image("old_projector.png", width=1024, height=768)
```

**Styling dla prezentacji:**
```python
fig.update_layout(
    font=dict(size=18),           # Większa czcionka (czytelność z daleka)
    title_font_size=28,           # Duży tytuł
    plot_bgcolor='white',         # Białe tło (projektor)
    paper_bgcolor='white',
    margin=dict(l=80, r=80, t=100, b=80)  # Marginesy
)
```

## Metoda 2: Interaktywny HTML w Prezentacji

### Jak to działa:
1. Zapisz wykres jako **HTML**
2. Otwórz HTML w przeglądarce
3. Prezentuj **przeglądarkę** (Alt+Tab między slajdami a HTML)

ALBO:

1. Zapisz wykres jako HTML
2. **Embed HTML w PowerPoint** (Insert → Object → Webpage)

### ✅ Plusy:
- Wykresy są **interaktywne** (hover, zoom, pan)!
- Możesz pokazać dane live podczas prezentacji
- Imponujące (publiczność może eksplorować dane)

### ❌ Minusy:
- Wymaga **przeglądarki** (nie działa w PDF)
- Embed w PowerPoint jest **niestabilny** (różne wersje Office)
- Trzeba Alt+Tab między slajdami a przeglądarką

### Kiedy używać:
- **Live demo** podczas prezentacji
- Chcesz pokazać **interaktywność** danych
- Prezentacja dla technicznej publiczności
- Masz dostęp do komputera i przeglądarki

---

### Kod - Export do HTML

````python
import plotly.express as px

df = px.data.gapminder().query("year == 2007")
fig = px.scatter(
    df, 
    x="gdpPercap", 
    y="lifeExp",
    size="pop", 
    color="continent",
    hover_name="country",  # Nazwa kraju po najechaniu
    title="Life Expectancy vs GDP (2007) - INTERAKTYWNY"
)

# EXPORT do standalone HTML
fig.write_html(
    "wykres_interaktywny.html",
    config={
        'displayModeBar': True,      # Pasek narzędzi (zoom, pan, save)
        'displaylogo': False,        # Ukryj logo Plotly
        'toImageButtonOptions': {    # Przycisk "Download as PNG"
            'format': 'png',
            'filename': 'custom_image',
            'height': 1080,
            'width': 1920
        }
    }
)

print("✅ Wykres zapisany jako wykres_interaktywny.html")
print("Teraz: Otwórz w przeglądarce i prezentuj!")
print("\nAlbo: PowerPoint → Insert → Text → Object → Webpage (niestabilne!)")
````

### Pro tip - Wiele wykresów na jednej stronie HTML:

Jeśli masz kilka wykresów, możesz zrobić **jedną stronę HTML** z wszystkimi:

````python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Utwórz subplot (2 wykresy obok siebie)
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Wykres 1: Life Expectancy", "Wykres 2: GDP Distribution")
)

df = px.data.gapminder().query("year == 2007")

# Wykres 1
fig.add_trace(
    go.Scatter(x=df['gdpPercap'], y=df['lifeExp'], mode='markers', 
               marker=dict(size=10, color=df['pop'], colorscale='Viridis')),
    row=1, col=1
)

# Wykres 2
fig.add_trace(
    go.Histogram(x=df['gdpPercap'], nbinsx=30),
    row=1, col=2
)

fig.update_layout(height=600, width=1400, title_text="Multi-Chart Dashboard")
fig.write_html("dashboard_prezentacja.html")

print("✅ Dashboard z 2 wykresami zapisany!")
````

## Metoda 3: Jupyter Notebook + RISE (Slajdy Live!)

### Jak to działa:
1. Tworzysz notatnik Jupyter z wykresami
2. Instalujesz **RISE** (Reveal.js extension)
3. Prezentacja **w przeglądarce** - interaktywne wykresy!

### ✅ Plusy:
- **Pełna interaktywność** (hover, zoom)
- Możesz **uruchomić kod** podczas prezentacji!
- Slajdy + kod + wykresy w jednym miejscu
- Idealne dla **technicznej publiczności**

### ❌ Minusy:
- Wymaga Jupyter Notebook + RISE
- Nie działa offline (bez Jupyter server)
- Mniej kontroli nad layoutem niż PowerPoint

### Kiedy używać:
- **Data science prezentacje**
- Chcesz pokazać **kod + wyniki**
- Live demo z wykonywaniem kodu
- Publiczność techniczna (programiści, analitycy)

---

### Setup RISE:

```bash

# Uruchom Jupyter
jupyter notebook
```

## Metoda 5: Dash App jako "Prezentacja"

### Jak to działa:
1. Tworzysz Dash app z wieloma sekcjami (jak slajdy)
2. Dodajesz przyciski "Next/Previous" lub tabs
3. Uruchamiasz Dash i prezentujesz przeglądarkę

### ✅ Plusy:
- **Pełna kontrola** - Python!
- Wszystko jest **interaktywne**
- Możesz dodać logikę (filtry, callbacks)
- Imponujące!

### ❌ Minusy:
- Musisz **programować** całą prezentację
- Brak gotowych templates (jak PowerPoint)
- Wymaga czasu na development

### Kiedy używać:
- **Interaktywne demo** produktu/analizy
- Chcesz że publiczność eksploruje dane
- Prezentacja wielokrotnego użytku (dashboard)
- Masz czas na development

---

### Przykład - Dash jako Prezentacja:

````python
from dash import Dash, html, dcc, Input, Output
import plotly.express as px

app = Dash(__name__)

# Dane
df = px.data.gapminder()

app.layout = html.Div([
    # Header
    html.Div([
        html.H1('Prezentacja Interaktywna - Gapminder Data',
                style={'textAlign': 'center', 'color': 'white'})
    ], style={'backgroundColor': '#2c3e50', 'padding': '20px'}),
    
    # Tabs = "Slajdy"
    dcc.Tabs(id='slides', value='slide1', children=[
        dcc.Tab(label='Slide 1: Overview', value='slide1'),
        dcc.Tab(label='Slide 2: Life Expectancy', value='slide2'),
        dcc.Tab(label='Slide 3: GDP Analysis', value='slide3'),
    ]),
    
    # Zawartość slajdu
    html.Div(id='slide-content', style={'padding': '40px'})
])

@app.callback(
    Output('slide-content', 'children'),
    Input('slides', 'value')
)
def render_slide(slide):
    if slide == 'slide1':
        return html.Div([
            html.H2('Welcome to Interactive Presentation!'),
            html.P('Navigate using tabs above. All charts are interactive - hover, zoom, pan!'),
            html.Img(src='https://via.placeholder.com/800x400?text=Intro+Slide')
        ])
    
    elif slide == 'slide2':
        fig = px.scatter(
            df[df['year'] == 2007],
            x='gdpPercap', y='lifeExp', size='pop', color='continent',
            title='Life Expectancy vs GDP (2007)',
            height=600
        )
        return html.Div([
            html.H2('Life Expectancy Analysis'),
            dcc.Graph(figure=fig)
        ])
    
    elif slide == 'slide3':
        fig = px.histogram(
            df[df['year'] == 2007],
            x='gdpPercap', color='continent',
            title='GDP Distribution by Continent',
            height=600
        )
        return html.Div([
            html.H2('GDP Analysis'),
            dcc.Graph(figure=fig)
        ])

if __name__ == '__main__':
    print("🚀 Prezentacja uruchomiona: http://127.0.0.1:8050/")
    print("💡 Tryb pełnoekranowy: F11")
    app.run(debug=True)
````

## 💡 Pro Tips - Prezentacje z Plotly

### 1. Czcionki większe niż normalnie:
```python
fig.update_layout(
    font=dict(size=18),        # Bazowa (normalnie 12)
    title_font_size=28,        # Tytuł (normalnie 16)
    legend_font_size=16        # Legenda
)
```

### 2. Kolory dla projektora (białe tło!):
```python
fig.update_layout(
    plot_bgcolor='white',      # Białe tło (lepiej na projektorze)
    paper_bgcolor='white'
)
```

### 3. Ukryj logo Plotly (profesjonalnie):
```python
fig.show(config={'displaylogo': False})
# lub
fig.write_html('chart.html', config={'displaylogo': False})
```

### 4. Dodaj watermark (branding):
```python
fig.add_layout_image(
    dict(
        source="https://your-logo.com/logo.png",
        xref="paper", yref="paper",
        x=1, y=1.05,
        sizex=0.2, sizey=0.2,
        xanchor="right", yanchor="bottom"
    )
)
```

### 5. Template dla prezentacji:
```python
# Utwórz własny template
presentation_template = dict(
    layout=dict(
        font=dict(size=18, family="Arial"),
        title_font_size=28,
        plot_bgcolor='white',
        paper_bgcolor='white',
        margin=dict(l=80, r=80, t=100, b=80)
    )
)

# Użyj dla wszystkich wykresów
fig = px.scatter(...)
fig.update_layout(presentation_template['layout'])
```

## ❓ FAQ

**Q: Czy mogę wstawić interaktywny wykres do PowerPointa?**  
A: Technicznie tak (Insert → Object → Webpage), ale jest **niestabilne**. Lepiej: PNG w slajdach + HTML w przeglądarce (Alt+Tab).

**Q: Jaki format najlepszy dla druku?**  
A: **SVG** (wektor - idealnie skaluje) lub **PNG** z wysoką rozdzielczością (scale=3).

**Q: Czy mogę animować slajdy w Plotly?**  
A: TAK! Użyj `px.scatter(..., animation_frame='year')` - wykres z timelinerem. Świetne do prezentacji!

**Q: Jak udostępnić interaktywną prezentację online?**  
A: 
- HTML na **GitHub Pages** (darmowe)
- Dash app na **Render** (darmowy tier)
- Jupyter notebook na **Binder** (darmowe, live)

**Q: Co jeśli chcę prezentację offline (bez internetu)?**  
A: 
- PNG → PowerPoint (zawsze działa)
- HTML standalone (plotly.js wbudowany w HTML)
- Reveal.js z lokalnymi plikami

**Q: Czy RISE działa w JupyterLab?**  
A: TAK, ale potrzeba extension: `jupyter labextension install @jupyterlab/rise`

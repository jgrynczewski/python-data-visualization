# Wczytywanie danych z pliku CSV

**CSV** (*Comma-Separated Values*) to najpopularniejszy format do przechowywania danych tabelarycznych.

**Dlaczego CSV?**
- Prosty format tekstowy (można otworzyć w notatniku)
- Uniwersalny - obsługiwany przez każde narzędzie do analizy danych
- Mały rozmiar pliku
- Łatwy do wygenerowania i parsowania

**Format:**
```
nazwa,kategoria,cena,ilosc
Laptop Dell XPS,Elektronika,4500.00,12
Mysz Logitech,Elektronika,89.99,45
...
```

Pierwszy wiersz to **nagłówki kolumn**, kolejne wiersze to **dane**.

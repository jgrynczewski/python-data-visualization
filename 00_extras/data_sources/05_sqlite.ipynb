# Wczytywanie danych z bazy SQLite

**SQLite** to lekka, bezserwerowa baza danych przechowywana w jednym pliku.

**Dlaczego SQLite?**
- Nie wymaga instalacji serwera bazy danych
- Cała baza w jednym pliku (`.db`)
- Idealna do aplikacji lokalnych, prototypów, testów
- Wspiera pełny SQL (SELECT, JOIN, WHERE, GROUP BY, etc.)
- Wbudowana w Python (moduł `sqlite3`)

**Kiedy używać:**
- Lokalne przechowywanie danych
- Aplikacje mobilne / desktopowe
- Prototypowanie przed migracją do większej bazy (PostgreSQL, MySQL)
